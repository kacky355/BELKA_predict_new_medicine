```python
import os

BASE_DIR = '/content/drive/MyDrive/BELKA_model'
OUT_DIR = os.path.join(BASE_DIR, 'SeXc_kfold')
if not os.path.exists(OUT_DIR):
    os.mkdir(OUT_DIR)
```

```python
import glob

subs = glob.glob(os.path.join(BASE_DIR, 'submission_*.csv'))
subs
```

```python
subs.pop(3)
```

```python
subs
```

```python
import pandas as pd
sub_dfs =[]
for sub in subs:
    col_name = sub.split('submission_')[-1].split('.')[0]
    df = pd.read_csv(sub, index_col='id').rename(columns={'binds' : f'binds_{col_name}'})
    sub_dfs.append(df)

sub_dfs[0].head()
```

```python
submission = pd.concat(sub_dfs, axis=1)
submission.head()
```

```python
submission['binds'] =submission.mean(axis=1)
```

```python
submission['binds'].to_csv(os.path.join(OUT_DIR, 'submission_SeXcv3_res_cnn.csv'))
```
